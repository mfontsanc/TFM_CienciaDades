## Fase 2: Comprensió de les dades

Els estudis clínics seleccionats com a fonts de dades es troben en el portal [clinicaltrials.gov](https://clinicaltrials.gov/). Aquest és mantingut per la U.S. National Library of Medicine, on s'inclouen assajos clínics tant dels Estatus Units com d'arreu del món, fent que sigui un dels portals públics d'accés gratuït més grans actualment disponibles.

El conjunt d'assajos clínics es poden descarregar com a fitxers XML en el següent enllaç:
* https://clinicaltrials.gov/api/gui/ref/download_all

La implementació del model ha de permetre torbar estudis clínics semblants per tal de crear una comunitat, on ha partir d'aquí es podran extreure la xarxa de col·laboradors. 


Per tal d'implementar el model per a trobar les comunitats, les dades que s'utilitzaran són les següents:
* Dades identificatives i descriptives de l'assaig.
* Dades sobre les condicions i els principals objectius.
* Dades sobre l'assaig clínic en la taxonomia [Mesh](https://www.nlm.nih.gov/mesh/meshhome.html).


Posteriorment, la xarxa de col·laboradors es podrà extreure de les següents seccions:
* Dades dels patrocinadors.
* Dades de les persones i organitzacions que hi treballen.


Finalment, la següent informació es farà ús per a la visualització de les dades:
* Dades sobre l'estat actual.
* Dades bàsiques sobre l'elecció dels participants.
